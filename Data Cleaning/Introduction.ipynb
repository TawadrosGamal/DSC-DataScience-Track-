### Secondly let's talk about the main tools every data scientist use:


### We have [Numpy](https://numpy.org/doc/stable/genindex.html) , [Pandas](https://pandas.pydata.org/docs/user_guide/index.html#user-guide) , and [MatPlotLib](https://matplotlib.org/3.3.1/contents.html)

### Usage of each of them:

* Pandas: is used to read different data types and load them into python, so that we can clean and adjust the data.
* Numpy: is mainly used in making matrices and perform quick and fast mathematical operations that is computationally expensive on basic lists/arrays, by using the memory much effectivly.
* MatPlotLib: is used in plotting the data and make good visualization to find pattern or present your data.
