### so what is broadcasting ?

```python
s+=2 #adds two to each item in s using broadcasting
s.head()
```

```python
s/=10 # devides each item in s using broadcasting
s.head()
```

```python
# you can also go in a for loop in each index and sum value +2
for index, value in s.items():
    s[index]= value+2
s.head()
```

```python
# reating 2 pandas series of sports-country 
# first one using the dictionary method
original_sports = pd.Series({'Archery': 'Bhutan',
                             'Golf': 'Scotland',
                             'Sumo': 'Japan',
                             'Taekwondo': 'South Korea'})

# second one using the lists method
cricket_loving_countries = pd.Series(['Australia',
                                      'Barbados',
                                      'Pakistan',
                                      'England'], 
                                   index=['Cricket',
                                          'Cricket',
                                          'Cricket',
                                          'Cricket'])

# appending both the pandas series together ,, remember the list.append --> s=[1,2,3]  s.append(4) xx s+4
all_countries = original_sports.append(cricket_loving_countries)
```
