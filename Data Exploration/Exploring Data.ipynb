## Creating dummy variables / one hot encoding 

### we want to create this to make use of the categorical col.
### we can make this by representing the categories of the columns by ones and zeros
#### representing male as 1 if found/zero if not found 
#### representing female the same way

```python
pd.get_dummies( df1['sex'] )
```

### but then the first col 'female' can predict the second col 'male',
### as if it's female then it's 0 in male, that's why it's DUMMY!
#### that's why one column is enough for us to use

```python
pd.get_dummies(df1['sex'],drop_first=True) # men = 1 , women = 0
```

```python
pd.get_dummies(df1, columns=['sex'], drop_first=True, prefix='sex') 
# you can put prefix before the categorical var to make it easy to distinguish
```

### splitting our data into features and labels for supervised learning task

```python
X= df1.drop(columns='survived') # creating features col without the servival col
```

```python
X.head()
```

```python
y=df1['survived'] # creating the labels col
```

```python
y.head()
```

## if splitting the data for unsupervised learning we discard the y col

## we can visualize and split the data for training and testing too

```python
X['age'].plot(kind='density')
```

## let's take 80% of the data as train and 20% as test

```python
size= int( len(X) * 80/100 )
size
```

```python
train_features = X.loc[ :size , : ] # all rows untill 1047
test_features = X.loc[size: , : ]   # all rows from 1047 untill the end

train_labels = y.loc[ :size ] # all rows untill the size - it's only 1 col
test_labels = y.loc[size: ]

print(f'train features head = \n{train_features.head()}\n\n\ntrain features length = \n{len(train_features)}\n')
print('-----------------------------------------------------------------------------------------------------------\n')
print(f'train labels head = \n{train_labels.head()}\n\ntrain labels length = \n{len(train_labels)}\n')
print('=============================================================================================================\n')
print(f'test features head = \n{test_features.head()}\n\n\ntest features length = \n{len(test_features)}\n')
print('-----------------------------------------------------------------------------------------------------------\n')
print(f'test labels head = \n{test_labels.head()}\n\ntest labels length = \n{len(test_labels)}\n')
```
